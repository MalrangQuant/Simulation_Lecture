```python
num = 10000
num_asset = 2
step = 32
T = 2 / step
dt = T #/num_time_step
r = 0.03 
q = 0
rho = 0.2
sigma1 = 0.3 #/ np.sqrt(step)
sigma2 = 0.4

s1_0 = 100
s2_0 = 100
rate = 0.125

path1 = np.ones((num, 1)) * s1_0
path2 = np.ones((num, 1)) * s2_0

z1 = np.random.normal(size=(step, num)).T
z2 = rho * z1 + np.sqrt(1-rho**2) * np.random.normal(size=(step, num)).T

x1 = np.exp( (r - q - 0.5 * (sigma1 **2)) * dt + sigma1 * np.sqrt(dt) * z1)
x2 = np.exp( (r - q - 0.5 * (sigma1 **2)) * dt + sigma1 * np.sqrt(dt) * z2)

for i in range(step):
    next_path = x1[:,i].reshape(num, 1) * path1[:, -1].reshape(num, 1)
    path1 = np.c_[path1, next_path]

for i in range(step):
    next_path = x2[:,i].reshape(num, 1) * path2[:, -1].reshape(num, 1)
    path2 = np.c_[path2, next_path]


path1.shape

check1_df = pd.DataFrame(data=np.round(path1,2), index=[x for x in range(num)], columns=[x for x in range(step+1)])
check2_df = pd.DataFrame(data=np.round(path2,2), index=[x for x in range(num)], columns=[x for x in range(step+1)])

check1 = pd.DataFrame(data=np.round(path1,2), index=[[x for x in range(num)],[0 for x in range(num)]], columns=[x for x in range(step+1)])
check2 = pd.DataFrame(data=np.round(path2,2), index=[[x for x in range(num)],[1 for x in range(num)]], columns=[x for x in range(step+1)])


df = pd.concat([check1, check2]).sort_index()
df

case_list = []

for idx in range(num):
    if (check1_df[8][idx] >= 85) and (check2_df[8][idx] >= 85):
        case = 1
    elif (check1_df[16][idx] >= 80) and (check2_df[16][idx] >= 80):
        case = 2
    elif (check1_df[24][idx] >= 75) and (check2_df[24][idx] >= 75):
        case = 3 
    elif (check1_df[32][idx] >= 70) and (check2_df[32][idx] >= 70):
        case = 4

    if ((check1_df.loc[idx] < 60).sum() >= 1) or ((check2_df.loc[idx] < 60).sum() >= 1):
        if (check1_df[32][idx]< 60) or (check2_df[32][idx]< 60):
            case = 6

    if ((check1_df.loc[idx] < 60).sum() < 1) and ((check2_df.loc[idx] < 60).sum() < 1):
        if (check1_df[32][idx]< 70) or (check2_df[32][idx]< 70):
            case = 5

    case_list.append(case)
    
print((sum(np.array(case_list) == 1) / num)*100)
print((sum(np.array(case_list) == 2) / num)*100)
print((sum(np.array(case_list) == 3) / num)*100)
print((sum(np.array(case_list) == 4) / num)*100)
print((sum(np.array(case_list) == 5) / num)*100)
print((sum(np.array(case_list) == 6) / num)*100)

```
